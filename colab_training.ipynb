# Cancer Type Training In Google Colab

Use this notebook to train the model in Colab. Each top-level folder becomes one class, and each class can contain only the scan or image-source folders that make sense for that cancer type:

```text
dataset/

    ct/

    pet/

  brain_cancer/
    mri/
      image1.dcm
  leukemia/

      image1.png
  skin_melanoma/
    dermoscopy/
      image1.jpg
```

This is an educational prototype, not a clinical diagnostic tool.

````python
from pathlib import Path
import os

candidates = [
    Path('/content/Cancer detection'),
    Path('/content/cancer-detection'),
    Path('/content'),
    Path.cwd(),
]

for candidate in candidates:
    if (candidate / 'train_model.py').exists():
        PROJECT_DIR = candidate
        break
else:
    raise FileNotFoundError('Could not find train_model.py. Upload the project folder to Colab first.')

os.chdir(PROJECT_DIR)
print('Project directory:', PROJECT_DIR)
````
